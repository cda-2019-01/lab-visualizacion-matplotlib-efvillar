```python
plt.style.use('seaborn-white')
fig, ax = plt.subplots(1, 6, sharex='col', sharey='row', figsize=(14, 7));
plt.subplots_adjust(wspace = 0, hspace=0.5)
for index, colname in enumerate(tabla1.columns):
    ax[index].scatter(list(tabla1.index), tabla1[colname], s=196,  marker='o', color = ['SteelBlue','SandyBrown','OliveDrab'])
    ax[index].set_xticklabels(tabla1.index, rotation=90 )
    ax[index].set_title(colname)
    fig.suptitle('Región', fontsize=20);
```
